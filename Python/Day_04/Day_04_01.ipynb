```python
#Creating a pandas series from a list 
s = pd.Series([10,20,30,40], name="My Numbers")
print(f"Pandas Series:\n{s}")
print(f"Type of Series:{type(s)}")
```

```python
#Creating a pandas DataFrame form a dictionary
data = { 'Name': ['Alice','Bob','Charlie','David'],
        'Age':[25,30,35,28],
        'City':['New York', 'London', 'Paris', 'New York'],
        'Salary':[200000,340000,450000,1200000]}
df = pd.DataFrame(data)
print(f"\nPandas DataFrame:\n{df}")
```
